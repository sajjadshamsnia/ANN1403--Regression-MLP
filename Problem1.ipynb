```python
# Sajjad Shamsnia
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('Dataset_1.csv')
species_num = []
for i in data['Species']:
    species_num.append(i)

### Part A
# species diversitys => sd 
sd = set(species_num)

# The number of each species => ns
ns = []
ns_dic = dict()
for s in sd:
    ns.append(species_num.count(s))
    ns_dic[s] = species_num.count(s)

y_axis = np.arange(0,60,5)
print(ns_dic)
print(sd)
print(ns)

plt.bar(list(sd),ns, color = 'green')
plt.title('Bar chart for each species')
plt.xlabel('Species')
plt.ylabel('Ù”Number')
plt.yticks(y_axis)
plt.grid(axis= 'y')
plt.show()
```

```python
# Relationship between size and weight
we = list()
c = list()
d = list() 
wi = list() 
v = list()
[we.append(i) for i in data['Weight']]
[c.append(i) for i in data['Cross Length']]
[d.append(i) for i in data['Diagnol Length']]
[wi.append(i) for i in data['Width']]
[v.append(i) for i in data['Vertical Length']]

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1)
fig.suptitle('A tale of 4 subplots')
fig.set_size_inches(10,10)
ax1.scatter(we,c)
ax1.set_ylabel('Cross Lenght')
ax2.scatter(we,d)
ax2.set_ylabel('Diagnol Length')
ax3.scatter(we,wi)
ax3.set_ylabel('Width')
ax4.scatter(we,v)
ax4.set_ylabel('Vertical Length')
ax4.set_xlabel('Weight')

plt.show()
```
